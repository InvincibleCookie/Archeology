```python
import ezdxf
import pandas as pd
from math import isclose

# Путь к файлу
file_path = '/content/РАСКОП 2.dxf'

# Координата X для фильтрации квадратов
x_threshold = 95282

# Список для хранения данных квадратов
squares_data = []

try:
    # Загрузка файла DXF
    doc = ezdxf.readfile(file_path)
    msp = doc.modelspace()

    # Список для хранения текстовых объектов
    text_objects = []

    # Шаг 1: Извлечение текстовых объектов
    for entity in msp:
        if entity.dxftype() == 'MTEXT':
            text_objects.append({
                'text': entity.dxf.text,
                'x': entity.dxf.insert[0],
                'y': entity.dxf.insert[1]
            })

    # Шаг 2: Извлечение квадратов (полигоны с 4 вершинами)
    for entity in msp:
        if entity.dxftype() == 'LWPOLYLINE' and entity.is_closed:
            points = entity.get_points("xy")
            if len(points) == 4 and all(p[0] > x_threshold for p in points):
                # Определяем центр квадрата
                center_x = sum(p[0] for p in points) / 4
                center_y = sum(p[1] for p in points) / 4

                # Проверяем, какой текстовый объект находится в этом квадрате
                square_number = "Unknown"
                for text_obj in text_objects:
                    text_x, text_y = text_obj['x'], text_obj['y']
                    if (min(p[0] for p in points) <= text_x <= max(p[0] for p in points) and
                        min(p[1] for p in points) <= text_y <= max(p[1] for p in points)):
                        square_number = text_obj['text']
                        break

                # Добавляем данные квадрата в список
                squares_data.append({
                    'Number': square_number,
                    'Point 1 (x, y)': points[0],
                    'Point 2 (x, y)': points[1],
                    'Point 3 (x, y)': points[2],
                    'Point 4 (x, y)': points[3],
                })

    # Преобразуем данные в DataFrame
    df = pd.DataFrame(squares_data)

    # Сохраняем в Excel
    output_path = '/content/squares_coordinates.xlsx'
    df.to_excel(output_path, index=False)
    print(f"Координаты квадратов сохранены в файл: {output_path}")

except Exception as e:
    print(f"Ошибка при обработке файла: {e}")
```

```python
import ezdxf

# Путь к файлу DXF
file_path = '/content/РАСКОП 2.dxf'

try:
    # Загрузка DXF-файла
    doc = ezdxf.readfile(file_path)
    msp = doc.modelspace()

    # Список для хранения текста
    all_texts = []

    # Извлекаем текстовые объекты (TEXT и MTEXT)
    for entity in msp:
        if entity.dxftype() == 'TEXT':  # Обычный текст
            all_texts.append(entity.dxf.text)
        elif entity.dxftype() == 'MTEXT':  # Многострочный текст
            all_texts.append(entity.text)

    # Извлекаем текст из блоков (если они есть)
    for block in doc.blocks:
        for entity in block:
            if entity.dxftype() == 'TEXT':
                all_texts.append(entity.dxf.text)
            elif entity.dxftype() == 'MTEXT':
                all_texts.append(entity.text)

    # Удаляем пустые строки и выводим текст
    all_texts = [text for text in all_texts if text.strip()]
    print("Собранный текст из DXF файла:")
    for text in all_texts:
        print(text)

except Exception as e:
    print(f"Ошибка при извлечении текста: {e}")
```
